```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
### Extract the Csv's in DataFrame ###
netflix_file = "../Resources/netflix_titles.csv"
netflix_df = pd.read_csv(netflix_file)
netflix_df.head()
```

```python
prime_file = "../Resources/Prime TV Shows Data set.csv"
prime_df = pd.read_csv(prime_file, encoding="ISO-8859-1")
prime_df.head()
```

```python
### Transform netflix and prime into Data Frame
```

```python
# Create filtered dataframe netflix #

netflix_cols = ["show_id", "title", "release_year"]
netflix_transformed = netflix_df[netflix_cols].copy()

# Rename the column names #

netflix_transformed = netflix_transformed.rename(columns={"show_id" : "id",
                                                         "title" :"Title",
                                                         "release_year" : "year"})
# Set index to start at id #

netflix_transformed.set_index("id", inplace=True)

# Print netflix

netflix_transformed.head()
```

```python
# Create Filtered Df prime #

prime_cols = ["S.no.", "Name of the Show", "Year of release"]
#prime_transformed = prime_df.rename(prime_cols).copy()

# Rename Columns #

prime_transformed = prime_df.rename(columns= {"S.no.": "id",
                                                      "Name of the show": "Title",
                                                       "Year of release"
                                                       :"year"})

# Set the index to id #

prime_transformed.set_index("id", inplace=True)

# Print prime #

prime_transformed.head()
```
